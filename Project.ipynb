```python
def transactions(path_to_data):
    Transactions = []
    with open(path_to_data,'r') as fd:
        for lines in fd:
            line = list(lines.strip().split())
            t = list(np.unique(list(map(int,line))))
            Transactions.append(t)
    return Transactions
```

```python
from collections import Counter

def generate_frequent_itemsets(result, min_support):
    item_count = Counter()
    
    for transaction in result:
        transaction = [str(item) for item in transaction]  
        for item in transaction:
            item_count[item] += 1
            
    frequent_itemsets = {item: count for item, count in item_count.items() if count >= min_support}
    all_frequent_itemsets = {item: count for item, count in frequent_itemsets.items()}
    k = 2
    
    while True:
        candidate_itemsets = []
        previous_frequent_itemsets = list(all_frequent_itemsets.keys())
        
        for i in range(len(previous_frequent_itemsets)):
            for j in range(i + 1, len(previous_frequent_itemsets)):
                candidate = sorted(previous_frequent_itemsets[i] + previous_frequent_itemsets[j])
                if len(candidate) == k and candidate not in candidate_itemsets:
                    candidate_itemsets.append(candidate)
        
        candidate_count = Counter()
        
        for transaction in result:
            transaction_set = set(transaction)
            for candidate in candidate_itemsets:
                if set(candidate).issubset(transaction_set):
                    candidate_count[tuple(candidate)] += 1
        
        frequent_itemsets = {list(itemset): count for itemset, count in candidate_count.items() if count >= min_support}
        
        if not frequent_itemsets:
            break
            
        all_frequent_itemsets.update(frequent_itemsets)
        k += 1
    
    return all_frequent_itemsets



min_support = 2
frequent_itemsets = generate_frequent_itemsets(result, min_support)

print("Frequent Itemsets:")
for itemset, count in frequent_itemsets.items():
    print(itemset, ":", count)
```

```python
from collections import Counter
import csv

def read_data(file_path):
    transactions = []
    with open(file_path, 'r') as file:
        for line in file:
            transaction = line.strip().split()
            transactions.append(transaction)
    return transactions

def apriori(result, min_support):
    k = 1
    item_count = Counter()
    for transaction in result:
        for item in transaction:
            item_count[str(item)] += 1
    L1 = {item: count for item, count in item_count.items() if count >= min_support}
    print(f"L1 (Support): {L1}")
    L_current = list(L1.keys())
    frequent_itemsets = L1.copy()
    while L_current:
        C_next = generate_candidates(L_current, k + 1)
        if not C_next:
            break
        candidate_count = count_support(result, C_next)
        L_next = {itemset: count for itemset, count in candidate_count.items() if count >= min_support}
        print(f"L{k + 1} (Support): {L_next}")
        frequent_itemsets.update(L_next)
        L_current = list(L_next.keys())
        k += 1
    return frequent_itemsets

def generate_candidates(L_current, k):
    candidates = []
    for i in range(len(L_current)):
        for j in range(i + 1, len(L_current)):
            candidate = sorted(set(L_current[i]) | set(L_current[j]))
            if len(candidate) == k:
                candidates.append(candidate)
    return candidates

def count_support(result, candidates):
    candidate_count = Counter()
    for transaction in result:
        transaction_set = set(map(str, transaction))
        for candidate in candidates:
            if set(candidate).issubset(transaction_set):
                candidate_count[tuple(candidate)] += 1
    return candidate_count

def compress_transactions(result, mapping):
    compressed_data = []
    for transaction in result:
        compressed_transaction = [mapping[item] for item in transaction if item in mapping]
        if compressed_transaction:
            compressed_data.append(compressed_transaction)
    return compressed_data

def calculate_compression_percentage(original, compressed):
    original_size = sum(len(transaction) for transaction in original)
    compressed_size = sum(len(transaction) for transaction in compressed)
    if original_size == 0:
        return 0
    compression_percentage = ((original_size - compressed_size) / original_size) * 100
    return compression_percentage

def write_to_csv(compressed_data, mapping, csv_filepath, min_support, compression_percentage):
    with open(csv_filepath, 'w', newline='') as csvfile:
        csv_writer = csv.writer(csvfile)
        csv_writer.writerow(["Mapping Table"])
        csv_writer.writerow(["Original Item", "Compressed Item"])
        for k, v in mapping.items():
            csv_writer.writerow([k, v])
        csv_writer.writerow([])
        csv_writer.writerow(["Compressed Transactions"])
        for transaction in compressed_data:
            csv_writer.writerow(transaction)
        csv_writer.writerow([])
        csv_writer.writerow(["Min Support Count"])
        csv_writer.writerow([min_support])
        csv_writer.writerow(["Compression Percentage"])
        csv_writer.writerow([compression_percentage])
def create_mapping(frequent_itemsets):
    mapping = {}
    unique_items = sorted(set(item for itemset in frequent_itemsets for item in itemset))
    
    groupings = {
        (1, 2, 3, 4): 'Group_A',
        (1, 2, 3, 5): 'Group_B',
        (1, 2, 3, 6): 'Group_C',
    }
    
    for idx, item in enumerate(unique_items):
        if item in mapping:
            continue
        
        for group, group_name in groupings.items():
            if all(str(i) in unique_items for i in group):
                mapping[group_name] = group
                for i in group:
                    mapping[str(i)] = group_name
    
    for idx, item in enumerate(unique_items):
        if item not in mapping:
            mapping[item] = f'Item_{idx + 1}'
    
    return mapping




def main():
    file_path = r"C:\python\DMDW_Project\demo.txt"
    result = read_data(file_path)
    min_support = 2  
    frequent_itemsets = apriori(result, min_support)
    mapping = create_mapping(frequent_itemsets)
    compressed_result = compress_transactions(result, mapping)
    compression_percentage = calculate_compression_percentage(result, compressed_result)
    print("\nCompressed Transactions:")
    for idx, transaction in enumerate(compressed_result):
        print(f"Compressed Transaction {idx + 1}: {transaction}")
    csv_filepath = r"C:\python\DMDW_Project\data.txt"
    write_to_csv(compressed_result, mapping, csv_filepath, min_support, compression_percentage)
    print("Compressed data and mapping table have been written to", csv_filepath)
    print(f"Compression Percentage: {compression_percentage:.2f}%")

if __name__ == "__main__":
    main()
```
